```python
#pull in website's source code
url = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/1'

page =requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup
```

```python
#find all
soup.find_all('tr',  attrs = {'class': 'colhead'})
```

```python
#find
#returns just one table header element that lists the column names from PLAYER to BA (Batting Average).
header = soup.find('tr',  attrs = {'class': 'colhead'})
header
```

```python
#were going to use Beautiful Soups .find_all() and find() methods to extract information from the source code.
#one player stat
row= soup.find('tr', attrs = {'class': 'oddrow player-10-33039'})
for data in row.find_all('td'):
    print(data.get_text())
```

```python
#create column names from the table header
header = soup.find('tr', attrs={'class': 'colhead'})

#extract col names
columns = [col.get_text() for col in header.find_all('td')]

final_df = pd.DataFrame(columns = columns)
final_df
```

```python
#pull in the player rows
#identify player rows
players = soup.find_all('tr', attrs={'class':re.compile('row player-10-')})
for player in players:
    
    #get stats for each player:
    stats = [stat.get_text() for stat in player.find_all('td')]
    
    #create a data frame for the single player's stats
    temp_df =pd.DataFrame(stats).transpose()
    temp_df.columns= columns
    
    #join the single player stats with overall data
    final_df = pd.concat([final_df, temp_df], ignore_index=True)
```
